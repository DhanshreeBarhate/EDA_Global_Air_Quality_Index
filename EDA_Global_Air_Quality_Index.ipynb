# LEVEL 3 ANALYSIS

```python
aql_overall_good = df[df['AQI Category'] == 'Good']
good = aql_overall_good.groupby('Country',as_index=False)['AQI Value'].mean().sort_values(by='AQI Value').head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='AQI Value', y='Country',data=good)
plt.title('Countries that have high Good Value')
```

```python
aql_overall_good = df[df['AQI Category'] == 'Moderate']
moderate = aql_overall_good.groupby('Country',as_index=False)['AQI Value'].mean().sort_values(by='AQI Value').head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='AQI Value', y='Country',data=moderate)
plt.title('Countries that have high Moderate Value')
```

```python
aql_overall_good = df[df['AQI Category'] == 'Unhealthy for Sensitive Groups']
sensitive_groups = aql_overall_good.groupby('Country',as_index=False)['AQI Value'].mean().sort_values(by='AQI Value').head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='AQI Value', y='Country',data=sensitive_groups)
plt.title('Countries that have high value in Unhealthy to Sensitive Groups category')
```

```python
aql_overall_good = df[df['AQI Category'] == 'Unhealthy']
unhealthy = aql_overall_good.groupby('Country',as_index=False)['AQI Value'].mean().sort_values(by='AQI Value').head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='AQI Value', y='Country',data=unhealthy)
plt.title('Countries that have high value in Unhealthy category')
```

```python
aql_overall_good = df[df['AQI Category'] == 'Very Unhealthy']
very_unhealthy = aql_overall_good.groupby('Country',as_index=False)['AQI Value'].mean().sort_values(by='AQI Value').head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='AQI Value', y='Country',data=very_unhealthy)
plt.title('Countries that have high value in Very Unhealthy category')
```

```python
aql_overall = df[['Country','City','AQI Value','AQI Category']]
aql_overall_haz = df[aql_overall['AQI Category'] == 'Hazardous']
hazardous = aql_overall_haz.groupby('Country',as_index=False)['AQI Value'].mean().sort_values(by='AQI Value',ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x='AQI Value', y='Country',data=hazardous)
plt.title('Most Country have worst overall AQI value')
```
